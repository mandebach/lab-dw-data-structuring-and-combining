```python
import pandas as pd
import numpy as np
```

```python
# Load datasets

url1 = 'file1_new.csv'
url2 = 'file2.csv'
url3 = 'file3.csv'

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
df1.info()
```

```python
df2.info()
```

```python
# 1. Standardize the column names of df2 by making lowercase, stripping white space and change ST to state

df2.columns = [column_name.lower() for column_name in df2.columns]
df2.columns = df2.columns.str.strip()
df2 = df2.rename(columns={'st': 'state'})
df2.columns
```

```python
for column in df2.columns:
    print(df2[column].value_counts())
    print()
```

```python
# 2. Clean the gender column

df2["gender"].unique()
df2["gender"] = df2["gender"].replace({
    "female": "F",
    "Male": "M"
})
df2["gender"].value_counts()
df2["gender"].isna().sum()
```

```python
# 3. Clean the state column
df2["state"].unique()
df2["state"] = df2["state"].replace({
    "AZ": "Arizona",
})
df2["state"].value_counts()
df2["state"].isna().sum()
```

```python
# 4. Clean the education column
df2["education"].unique()
df2["education"] = df2["education"].replace({"Bachelors": "Bachelor"})
df2["education"].value_counts()
df2["education"].isna().sum()
```

```python
# 5. Clean the customer lifetime value column
df2["customer lifetime value"] = pd.to_numeric(
    df2["customer lifetime value"].str.rstrip("%"),
    errors="coerce"
)

# Convert percentage-formatted values back to their original scale
df2["customer lifetime value"] /= 100

df2["customer lifetime value"].head(10)

df2["customer lifetime value"].dtype
```

```python
# 6. Clean Vehicle column

df2["vehicle class"].unique()
df2["vehicle class"] = df2["vehicle class"].replace({
    "Luxury SUV": "Luxury",
    "Sports Car": "Luxury",
    "Luxury Car": "Luxury",
})
df2["vehicle class"].value_counts()
```

```python
# 7. Clean number of open complaints column
df2["number of open complaints"].value_counts()

# Split it into a list, convert to number, take the middle value only

df2["number of open complaints"] = pd.to_numeric(
    df2["number of open complaints"].str.split("/").str[1],
    errors="coerce"
)
```

```python
df2["number of open complaints"].value_counts()
```

```python
df3.info()
```

```python
# 1. Standardize column names to be lowercase and remove spaces
df3.columns = [column_name.lower() for column_name in df2.columns]
df3.columns = df3.columns.str.strip()
df3.info()
```

```python
for column in df3.columns:
    print(df3[column].value_counts())
    print()
```

```python
# 2. Clean Vehicle column

df3["vehicle class"].unique()
df3["vehicle class"] = df3["vehicle class"].replace({
    "Luxury SUV": "Luxury",
    "Sports Car": "Luxury",
    "Luxury Car": "Luxury",
})
df3["vehicle class"].value_counts()
```

```python
df3.isna().sum()
```

```python
# Concat all three clean datasets into a new df

df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df.info()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Load dataset

marketing_url = 'marketing_customer_analysis_clean.csv'

df4 = pd.read_csv(marketing_url)
```

```python
df4.info()
```

```python
# Check each column for data cleaning

for column in df4.columns:
    print(df4[column].value_counts())
    print()
```

```python
# Columns to clean: number_of_open_complaints

df4["number_of_open_complaints"].value_counts()
df4.loc[df4["number_of_open_complaints"] == 0.384256]

df4["number_of_open_complaints"].unique()

# Find out how many rows are affected
df4[df4["number_of_open_complaints"] == 0.38425611].shape

# Find which ones are not an integer and make them a missing value to deal with corrupted 0.38425611
df4.loc[
    ~np.isclose(df4["number_of_open_complaints"] % 1, 0),
    "number_of_open_complaints"
] = np.nan
```

```python
df4["number_of_open_complaints"].value_counts().sort_index()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Pivot table that shows customer lifetime value per sales channel

result = pd.pivot_table(
    df4,
    values="customer_lifetime_value",
    index="sales_channel",
    aggfunc="sum"
).round(2)

print(result)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot = pd.pivot_table(
    df4,
    values="customer_lifetime_value",
    index="gender",
    columns="education",
    aggfunc="mean"
).round(2)

print(pivot)
```

# Insights

1. Women with an education level or high school or below seem to spend the most overall
2. Men with a Master level education spend the most in the male category
3. Women and men with a doctor level education spend the least.
